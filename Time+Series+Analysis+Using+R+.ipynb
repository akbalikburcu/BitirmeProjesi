# TIME SERIES ANALYSIS

The main purpose of the time series analysis is forecasting. Time series data can have a wide variety of models and it is useful to categorize some of the models and behaviors that can be seen in the time series.For this reason, a method has been developed to separate time series into a **_trend_**, a **_seasonal_**, **_a cyclical_** and an **_irregular component_**.

## Time Series Components

### Trend

1. There is a trend when there is a prolonged increase or decrease in the data.
2. This may be an increasing or decreasing trend. 
3. It does not have to be linear.

### Seasonal

1. If the data we examine is affected by seasonal factors, a seasonal pattern is formed. 
2. Seasonal is always a fixed and known period.

### Cyclical

1. When the data are rising and falling, that is, when there is no fixed period, a cyclic pattern occurs. 
2. These fluctuations usually long at least 2 years.

### Irregular Component 

1. The irregular component is stochastic and our aim is to make a forecast using this component.

# Time Series Decomposition

We consider that the $Y_t$ time series consists of three components: a seasonal component, a component of a trend loop, and the remaining component. We can express this model in the following way,

$$ y_t = S_t + T_t + E_t $$

_**Additive Model**: A data model in which the effects of individual factors are differentiated and added together to model the data. An additive model is optional for decomposition procedures._

Where $y_t$ is the t-period value, $S_t$ is the seasonal component at time t, $T_t$ is the component of the trend cycle at time t, and $E_t$ is the remaining component at time t. Alternatively, a multiplicative model is written as follows:

$$ y_t = S_t \times T_t \times E_t$$

_**Multiplicative Model**: This model assumes that as the data increase, so does the seasonal pattern. Most time series plots exhibit such a pattern. In this model, the trend and seasonal components are multiplied and then added to the error component._

If the size of the seasonal fluctuations or the variation around the trend cycle does not depend on the level of the time series, the additive model is most appropriate. A multiplicative model is more suitable if the variation around the seasonal pattern or the variation around the trend line is proportional to the level of the time series. Multiplicative models have become widespread with economic time series.

An alternative way of using a multiplicative model is to transform the data first as far as the variant in the heave is stable over time, and then use an additive model. When using a day conversion, it is equivalent to using a multiplicative decay
$y_t = S_t \times T_t \times E_t$, equivalent to $logy_t = logS_t + logT_t + logE_t$.

# STL Decomposition

STL, "Seasonal and Trend Decomposition using Loess" is a method used to estimate nonlinear relationships. It decomposes a time series into seasonal, trend and irregular components using loess. The methodology was presented by Robert Cleveland, William Cleveland, Jean McRae and Irma Terpenning in the Journal of Official Statistics in 1990.

STL has some advantages over the classical decomposition method:

1. STL does not only have a certain frequency, it works for every kind of seasonality.
2. The seasonal component is allowed to change over time and the rate of change can be controlled by the user.
3. The smoothness of the power cycle can also be controlled by the user.

Because of that STL decomposition works with only additive time series, we should take logs of gold data to obtain a multiplicative decomposition of it.

## Autocorrelation Function and Partial Autocorrelation Function Plots

We are able to identify the number of AR and MA terms that we will need later by looking at the (ACF) and (PACF) plots of the differing series.


### ACF

Autocorrelation is the linear correlation of a signal with itself at two different points in time, ACF (autocorrelation function) is just such correlation as a function of the lag h  between two points of time, like acf(h)=corr($ x_t $ ,$ x_t+_h $). The ACF for a time series $y_t$ , _t = 1,...,N_, is the sequence $ρ_h$ , _h = 1, 2,...,N – 1_.   

### PACF

Partial autocorrelation function is essentially the autocorrelation of a signal with itself at different points in time, with linear dependency with that signal at shorter lags removed, as a function of lag between points of time. Informally, the partial correlation between $x_t$ and $x_t+_h$ is the autocorrelation between $x_t$ and $x_t+_h$ without the contribution of $ x_t+_1 ,x_t+_2 ,....,x_t+_h−1 $.The PACF is the sequence $ϕ_h,_h$, _h = 1, 2,...,N – 1_.
